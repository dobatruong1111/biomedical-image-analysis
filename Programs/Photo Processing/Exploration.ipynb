# Shape, sampling and field of view
<p> - Shape (hình dạng): số lượng pixel trên mỗi trục của đối tượng </p>
<p> - Samping (tỷ lệ lấy mẫu): Kích thước của mỗi pixel trong thực tế (mm) hay Khoảng cách hai pixel trong thực tế </p>
<p> - Field of view (Trường nhìn): Kích thước thực của đối tượng (mm) = shape[i] * sampling[i] </p>

```python
#Shape
n0, n1, n2 = vol2.shape #: vd 20 anh moi anh (256,256) => (n0 = 20, n1 = 256,n2 = 256)
#Sampling
d0, d1, d2 = vol2.meta['sampling']
#Field of view
fov = n0*d0, n1*d1, n2*d2
print(fov)
```

# Generating Subplots
<p> - Chúng ta có thể vẽ nhiều hình ảnh trong một hình để khám phá dữ liệu một cách nhanh chóng. Sử dụng plt.subplots() để tạo ra một mảng các subplots. </p>

```python
fig, axes = plt.subplots(nrows = 1,ncols = 3,figsize = (15,5))
axes[0].imshow(vol2[0],cmap = 'gray')
axes[1].imshow(vol2[1],cmap = 'gray')
axes[2].imshow(vol2[2],cmap = 'gray')

for ax in axes:
    ax.axis('off')
    
plt.show()
```

# Slice 3D Images
<p> - Cách đơn giản nhất để vẽ hình ảnh 3D và 4D là bằng cách cắt chúng thành nhiều khung hình 2D. Vẽ nhiều lát cắt tuần tự có thể giúp bạn hiểu toàn bộ hình ảnh. </p>
<p> - Bất kỳ hai chiều nào của một mảng đều có thể tạo thành một hình ảnh và việc cắt dọc theo các trục khác nhau có thể cung cấp một góc nhìn hữu ích. Tuy nhiên, tỷ lệ lấy mẫu không đồng đều có thể tạo ra hình ảnh bị bóp méo. </p>
<p> - Thay đổi tỷ lệ khung hình có thể giải quyết vấn đề này bằng cách tăng chiều rộng của một trong các kích thước. Và trong hàm imshow() của thư viện Matplotlib có cung cấp một đối aspect giúp điều chỉnh tỷ lệ khung hình. </p>

```python
view_1v2 = vol2[10,:,:] # mp axial: lát cắt cho ta góc nhìn từ trên xuống
view_0v2 = vol2[:,128,:] # mp coronal: lát cắt cho ta góc nhìn nhìn đối diện 
view_0v1 = vol2[:,:,128] # mp sagittal: lát cắt cho ta góc nhìn một bên 
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 3, figsize = (15,5))
axes[0].imshow(view_1v2, cmap = 'gray',aspect = d1/d2)
axes[1].imshow(view_0v2, cmap = 'gray',aspect = d0/d2)
axes[2].imshow(view_0v1, cmap = 'gray',aspect = d0/d1)

for ax in axes:
    ax.axis('off')
plt.show()
```
